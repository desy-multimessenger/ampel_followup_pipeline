```python
import logging
from nuztf.skymap_scanner import SkymapScanner
```

```python
logger = logging.getLogger("nuztf")
logger.setLevel(logging.INFO)
```

```python
# Look for transients first detected within 72 hours of merger
n_days = 3.0
```

GravWaveScanner can be initialised with optional arguments *gw_name* and *rev*. By default, it loads the most recent LIGO public event, and the most recent revision. Alternatively, a fits file can be saved to the LIGO_skymaps directory, and its name specified with *gw_file*.

```python
prob_threshold = 0.95
n_days = 3

# skymap_file = "IGWN-GWTC3p0-v1-GW200210_092254_PEDataRelease_cosmo_reweight_C01_Mixed.fits"
# skymap_file="GRB220617A_IPN_map_hpx2.fit"
skymap_file="https://heasarc.gsfc.nasa.gov/FTP/fermi/data/gbm/triggers/2023/bn230511945/quicklook/glg_healpix_all_bn230511945.fit"
```

```python
scanner = SkymapScanner(
    event=skymap_file,
    prob_threshold=prob_threshold,
    n_days=n_days
)
```

```python
scanner.plot_skymap()
```

```python
scanner.get_alerts()
scanner.filter_alerts()
scanner.export_cache_to_fritz() 
```

```python
scanner.create_candidate_summary()
scanner.create_overview_table()
```

```python
scanner.export_cache_to_fritz() 
```

```python
scanner.plot_coverage()
```

```python
print(scanner.draft_gcn())
```
